```python
def radian(x):
    return x / 180 * 3.1415

for x in range(0, 360, 90):
    print('도: {}, 라디안: {:.2f}'.format(x, radian(x)))
```

```python
lambda_radian = (lambda x:x / 180 * 3.1415)

for x in range(0, 360, 90):
    print('도: {}, 라디안: {:.2f}'.format(x, lambda_radian(x)))
```

```python
class HelloClass:
    def __init__(self, msg):
        self.msg = msg

    def output(self):
        print(self.msg)

hello = HelloClass('Hello World')
hello.output()
```
